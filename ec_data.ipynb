```python
# Get column IDs and labels for data of interest from external file
column_ids = []
with open('metadata.txt') as f:
    for line in f:
        column_ids.append(tuple(line.strip('\n').split(',', 1)))

for labels in column_ids:
    print(f'{labels[0]}: {labels[1]}')
```

```python
# https://factfinder.census.gov/bkmk/table/1.0/en/ACS/17_1YR/DP05/0100000US|0100000US.04000
census_df = pd.read_csv('ACS_17_1YR_DP05/ACS_17_1YR_DP05_with_ann.csv', 
                        header=0,
                        index_col=0,   # Use state names as index
                        usecols=[i[0] for i in column_ids])

# Sorry Puerto Rico...
census_df.drop('Puerto Rico', inplace=True)
census_df

```

```python
ec_df = pd.read_html('https://www.archives.gov/federal-register/electoral-college/allocation.html',
                     match='Number of Electoral Votes',
                     index_col=0)[2]  # Use states as index
ec_df
```

```python
# Save dataframes to csv files
census_df.to_csv('df_census.csv')
ec_df.to_csv('df_ec.csv')
```
